#Removing Unecessary Columns

```python
df.drop(['Salary','Salary Currency','Employee Residence'],axis = 1 , inplace = True)
```

```python
df.head()
```

```python
## Converting Salaries In Indian rupees For better understandment
Salary_in_INR = []
for index , row in df.iterrows():
    if row['Year'] == 2020:
        new_salary = row['Salary in USD']*72.85
    elif row['Year'] == 2021:
        new_salary = row['Salary in USD']*73.86
    elif row['Year'] == 2022:
        new_salary = row['Salary in USD']*80.50
    else:
        new_salary = row['Salary in USD']*83.94
    Salary_in_INR.append(new_salary)
```

```python
## Creating a New column 
df['Salary In INR'] = Salary_in_INR
df
```

Or Second Way To Acheive the same result

```python
Salary_in_inr = []
for salary,year in zip(df['Salary in USD'],df['Year']):
    if year == 2020:
        new_sal = salary *72.85
        Salary_in_inr.append(new_sal)
    elif year == 2021:
        new_sal = salary * 73.86
        Salary_in_inr.append(new_sal)
    elif year == 2022:
        new_sal = salary * 80.50
        Salary_in_inr.append(new_sal)
    else:
        new_sal = salary * 83.94
        Salary_in_inr.append(new_sal)
Salary_in_inr[0:5]  ## Taking Overview of My List
```

```python
## Making a New Column using Salary_in_inr
df['Salary_in_inr'] = Salary_in_inr
df
```

```python
## Now we obviously don't need two salary_in_inr column So i am dropping any one
df.drop(['Salary_in_inr'],axis = 1 ,inplace = True)
```

## Increament In Jobs Year vise

```python
year_vise_data = df.groupby(['Job Title','Year'])['Salary In Lacs'].mean().reset_index()
```

```python
year_vise_data = year_vise_data[year_vise_data['Job Title'].isin(['Data Scientist','Data Engineer','Data Analyst'])]
```

```python
year_vise_data
```

```python
## Let's Find Percentage Increment In Salry Of Each Group by Year

all_data = []
for data in zip(year_vise_data['Salary In Lacs'], year_vise_data['Job Title']):
   all_data.append(data)
all_data
diff_percentage = [0]

prev_job = all_data[0][1]
prev_salary  = all_data[0][0]

for i in range(1, len(all_data)):
    curr_job = all_data[i][1]
    cur_salary = all_data[i][0]

    if curr_job == prev_job:
        # print(f"Difference : { cur_salary - prev_salary}")
        diff = cur_salary - prev_salary
        percentage = round((diff / cur_salary) * 100, 2)
        diff_percentage.append(percentage)
        
    if curr_job != prev_job:
        diff_percentage.append(0)
        

    prev_job, prev_salary = curr_job, cur_salary

year_vise_data['Percentage Change'] = diff_percentage
```
